### Task - 01

**Show graph between all countries developer and python programming languge they uses

```python
country_grp = df.groupby('Country')
```

```python
country_grp['LanguageWorkedWith'].value_counts()
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

***Total number of respondant from each country

```python
total_respondant = df['Country'].value_counts()
total_respondant
```

***Programming languge known by each country

```python
python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
python_df = pd.concat([total_respondant , python] , axis='columns' , sort=False)
graph_data = python_df.head(10)
graph_data
```

```python
graph_data.rename(columns={'Country':'Total User' , 'LanguageWorkedWith':'Python User'},inplace=True)
```

***Adding Java Data

```python
Java = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Java').sum())
Java
```
